```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import cv2

from glob import glob
from os import system,path,mkdir

%matplotlib inline
```

```python
PATH = "./images/original/"
people = [path.split(path.abspath(i))[-1] for i in glob(f"{PATH}/*")]
```

```python
haar = cv2.CascadeClassifier()
haar.load("../data/haarcascade_frontalface_default.xml")
```

```python
!rm -rf "./images/train"
!mkdir "./images/train"

i = 0
j = 0
n = 150
while j < n: 
    p = people[i]
    imgs = glob(f"./images/original/{p}/*.jpg")
    mkdir(f"./images/train/{p}")
    try:
        for k in range(3):
            img = cv2.imread(imgs[k])
            x,y,w,h = haar.detectMultiScale(img)[-1]
            cv2.imwrite(
                f"images/train/{p}/{k}.jpg",
                cv2.resize(
                    img[y-3:y+h+3,x-3:x+w+3],
                    (128,128),
                    interpolation=cv2.INTER_AREA
                )
            )
        j += 1
        i += 1
        print (f"\r{j}/{n}")
    except:
        system(f"rm -rf images/train/{p}")
        i += 1
    
```
